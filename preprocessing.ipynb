```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import numpy as np
```

```python
df = pd.read_csv('data/GlobalWeatherRepository.csv')
df.head()
```

```python
print(df.dtypes)
```

```python
# Convert date-time columns
df['last_updated'] = pd.to_datetime(df['last_updated'])
```

```python
# Define features and target
features = df[['latitude', 'longitude', 'temperature_celsius', 'wind_mph', 'pressure_mb', 'humidity', 'cloud',
               'feels_like_celsius', 'visibility_km', 'uv_index', 'gust_mph', 'air_quality_Carbon_Monoxide',
               'air_quality_Ozone', 'air_quality_Nitrogen_dioxide', 'air_quality_Sulphur_dioxide', 'air_quality_PM2.5',
               'air_quality_PM10', 'air_quality_us-epa-index', 'air_quality_gb-defra-index']]
target = df['temperature_celsius']  # For regression example
```

```python
numerical_cols = features.columns.tolist()
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
    ])
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
# Create a pipeline
pipeline = Pipeline(steps=[('preprocessor', preprocessor)])
```

```python
# Preprocess the data
X_train = pipeline.fit_transform(X_train)
X_test = pipeline.transform(X_test)
```

```python
# Save preprocessed data
np.save('X_train.npy', X_train)
np.save('X_test.npy', X_test)
np.save('y_train.npy', y_train)
np.save('y_test.npy', y_test)
```
